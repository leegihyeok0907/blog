```python
import geopandas as gpd
import pandas as pd
```

```python
# Read in the data
full_data = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/데이터마이닝/geospatial learn course data/DEC_lands/DEC_lands/DEC_lands.shp")

# View the first five rows of thee data
full_data.head()
```

```python
type(full_data)
```

```python
data = full_data.loc[:, ["CLASS", "COUNTY", "geometry"]].copy()
```

```python
# How many lands of each type are there?
data.CLASS.value_counts()
```

```python
# Select lands that fall under the "WILD FOREST" or "WILDERNESS" category
wild_lands = data.loc[data.CLASS.isin(['WILD FOREST', 'WILDERNESS'])].copy()
wild_lands.head()
```

```python
wild_lands.plot()
```

```python
# View the first five entries in the "geometry" column
wild_lands.geometry.head()
```

```python
# Campsites in New York state (Point)
POI_data = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/데이터마이닝/geospatial learn course data/DEC_pointsinterest/DEC_pointsinterest/Decptsofinterest.shp")
campsites = POI_data.loc[POI_data.ASSET=='PRIMITIVE CAMPSITE'].copy()

# Foot trails in New York state (LineString)
roads_trails = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/데이터마이닝/geospatial learn course data/DEC_roadstrails/DEC_roadstrails/Decroadstrails.shp")
trails = roads_trails.loc[roads_trails.ASSET=='FOOT TRAIL'].copy()

# County boundaries in New York state (Polygon)
counties = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/데이터마이닝/geospatial learn course data/NY_county_boundaries/NY_county_boundaries/NY_county_boundaries.shp")
```

```python
# Define a base map with county boundaries
ax = counties.plot(figsize=(10,10), color='none', edgecolor='gainsboro', zorder=3)

# Add wild lands, campsites, and foot trails to the base map
wild_lands.plot(color='lightgreen', ax=ax)
campsites.plot(color='maroon', markersize=2, ax=ax)
trails.plot(color='black', markersize=1, ax=ax)
```
